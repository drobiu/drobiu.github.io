```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../data/eeg-lab-example-no-transpose-all.csv')
data = data.transpose()
data.columns = data.iloc[0]
data = data.iloc[1:]
data = data.reset_index()
columns = list(data.columns)
columns[0] = 'time'
data.columns = columns
data = data.astype(float)
```

```python
data
```

```python
json_data = [{'name': k, 'data': list(data.loc[:, k].values)} for k in data.columns]
```

```python
import json
```

```python
with open('../data/eeg.json', 'w') as json_file:
    json.dump(json_data, json_file)
    json_file.close()
```
